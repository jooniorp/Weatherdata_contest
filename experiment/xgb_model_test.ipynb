```python
import pandas as pd

df_test = pd.read_csv("electric_test.csv")
df_test.head()
```

```python
col_names = list(df_test.columns)
col_names
```

```python
# 월, 일 변수 추가
df_copy = df_test.copy()

df_copy['TM'] = pd.to_datetime(df_copy['TM'])

df_copy['month'] = df_copy['TM'].dt.month
df_copy['day'] = df_copy['TM'].dt.day
```

```python
# 카테고리화 할 변수
features = ['HH24','weekday','week_name', 'month','day']

#모델 검증을 위해 카테고리형 변수로 변환
for f in features:
    df_copy[f] = df_copy[f].astype('category')

df_copy.info()
```

```python
df_copy['week_name'] = df_copy['week_name'].astype('int')
```

```python
df_copy.head()
```

```python
# 카테고리형 변수 원-핫 인코딩
df_copy = pd.get_dummies(df_copy, columns=['HH24','weekday','week_name', 'month','day'])
```

```python
import joblib

# 모델 로드
loaded_model = joblib.load('xgb_model.pkl')
```

```python
df = df_copy.drop(['elect','NUM','TM'], axis=1)
```

```python
df.columns = df.columns.str.lower()
```

```python
saved_features = joblib.load('xgb_model_features.pkl')
```

```python
df.columns
```

```python
df = df[saved_features]
```

```python
y_pred = loaded_model.predict(df)
```

```python
len(y_pred)
```

```python
y_pred
```

```python
df_test['elect'] = y_pred
```
